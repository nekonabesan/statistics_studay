### 設問１


導出の際に会社$ A $に対するスワップの価値は$ V_{A} = X - NX\sum_{t=0}^{T} d(0, t) $であることを思い出して下さい。

ここで第一項の$ X $はなんですか？

1. $ N $
2. $ N(d(0, T) - 1) $
3. $ N(1 - d(0, T)) $




### 回答１

**正解：3. $N(1 - d(0, T))$**

**説明**：
スワップ価値式
\[
V_A = X - N\sum_{t=1}^{T} d(0,t)\,c
\]
（授業ノートの記法に合わせると固定レッグと変動レッグの差）において、$X$ は**変動レッグ（フローティングレッグ）の現在価値**を表します。標準的な金利スワップの評価では、この部分は
\[
X = N(1-d(0,T))
\]
となります。

**補足**：
$d(0,T)$ は満期 $T$ の割引因子です。したがって $1-d(0,T)$ は「現在から満期までの累積割引効果」に対応し、元本 $N$ を掛けることで変動レッグの現在価値になります。上の Python セルでは、この対応を式として確認できるようにしています。

```python
# 設問1の計算過程（数値例）
# 変動レッグの現在価値 X = N(1 - d(0,T)) を確認する

N = 100_000_000  # 元本 1億
r = 0.02         # 年率2%
T = 5            # 5年

# 単純化のため連続複利で割引因子を作る
import math
d_0_T = math.exp(-r * T)

X = N * (1 - d_0_T)

print(f"d(0,T) = {d_0_T:.6f}")
print(f"X = N(1-d(0,T)) = {X:,.0f}")
print("→ 設問1の正解は 3. N(1 - d(0,T))")
```

### 設問２

リスクの観点から金融仲介業者とスワップを締結する理由は何ですか？

1. 買い手がデフォルトリスクにさらされないようにするため
2. 両当事者がデフォルトリスクにさらされないようにするため
3.  売り手がデフォルトリスクにさらされないようウにするため





### 回答２

**正解：2. 両当事者がデフォルトリスクにさらされないようにするため**

**説明**：
金融仲介業者（ディーラーや清算機関）を介在させると、取引当事者同士が直接信用リスクを取り合う構造を弱められます。実務では証拠金管理やネット決済などの仕組みにより、**売り手・買い手の双方のカウンターパーティリスク**を低減することが目的です。

**補足**：
実際にはリスクが「ゼロ」になるわけではなく、仲介業者自体の信用力・証拠金制度・清算ルールに依存します。そのため、取引先の信用評価と担保管理は継続的に必要です。

### 設問３

金融取引において取引所が望ましくない理由は何ですか？

1. 市場に多数の商品があるようにします
2. 価格が透明であることを確認して下さい
3. 需給の一致を解決
4. 個人が直面するデフォルトリスクの低減


### 回答３

**正解：1. 市場に多数の商品があるようにします**

**説明**：
取引所取引は標準化が前提なので、商品が増えすぎると流動性が分散し、板が薄くなりやすくなります。したがって「市場に多数の商品があるようにする」ことは、取引所の効率性という観点では望ましくありません。

**補足**：
選択肢2〜4（価格透明性、需給マッチング、デフォルトリスク低減）は、一般に取引所の代表的な利点です。OTC は柔軟性が高い一方で、取引所ほど標準化メリットを得にくいというトレードオフがあります。

### 設問４

次のうちで先物を使った完全なヘッジが不可能なケースはどれですか？

※該当するものすべてを選択

1. 原資産の先物はありません
2. 完全なヘッジを行うには端数評価された先物が必要ですが、これは実際には利用できません
3. ペイオフはスポットに対して非線形です








### 回答４

**正解：1, 2, 3（すべて該当）**

**説明**：
先物で完全ヘッジするには、ヘッジ対象リスクと同一（または完全相関）で、必要数量を連続的に調整できる取引手段が必要です。以下はいずれも完全ヘッジを妨げます。

- 1: 原資産に対応する先物が存在しない（クロスヘッジとなりベーシスリスクが残る）
- 2: 必要契約数が端数になるが、実際には整数契約しか取引できない（離散化誤差）
- 3: ペイオフがスポットに対して非線形（先物の線形ペイオフだけでは完全複製できない）

**補足**：
3 の典型例はオプションです。非線形ペイオフを先物だけで完全再現するには連続リバランスなど理想化条件が必要で、実務では近似ヘッジ（デルタヘッジ等）になります。
