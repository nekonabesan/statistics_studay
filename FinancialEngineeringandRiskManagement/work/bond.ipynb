### 設問１

$ f_{uv} $すなわち、今日のフォワードレートと明日のフォワードレートは同じですか？

- フォワードレートは時間経過とともに変化します
- はい、フォワードレートは長期間変化しません



### 回答１

**正解:** 「フォワードレートは時間経過とともに変化します」

**説明:**
一般に、フォワードレート $f_{u,v}$ はその時点でのゼロクーポン価格（割引因子）から決まる市場整合的なレートです。市場金利・期待・需給が変われば、翌日には割引曲線が変化し、フォワードレートも変わります。したがって「今日のフォワードレートと明日のフォワードレートは同じ」とは限りません。

**補足:**
同一の将来区間 $(u,v)$ を見ていても、観測時点が変わると再評価されるため、値は通常変動します。

### 設問２

満期が$ T $、額面が$ F $の先渡契約である場合、商品のスポット価格$ S_{T} $、$ T $時点の先渡契約の価値$ f_{T} $は？

1. $ S_{T} - F $
2. $ F - S_{T} $
3. $ S_{T} $
4. $ F $




### 回答２

**正解:** 1. $S_T - F$

**説明:**
満期 $T$ 時点での先渡契約（ロング側）の価値は、受け取る原資産の市場価値 $S_T$ から、支払う受渡価格（契約で固定した価格）$F$ を引いた
$$
f_T = S_T - F
$$
で与えられます。

**補足:**
ショート側（売り手）の価値は符号が逆になり、$F - S_T$ です。

### 設問３

固定金利$ (r) $があると仮定した場合、満期先物契約$ (T) $の額面はいくらですか？

1. $ \frac{S_{T}}{(1 + r)^{T}} $
2. $ \frac{S_{0}}{r} $
3. $ S_{0}(1 + r)^{T} $
4. $ S_{T} $





### 回答３

**正解:** 3. $S_0(1+r)^T$

**説明:**
固定金利 $r$（年複利）を仮定すると、満期 $T$ の理論先渡価格（受渡価格）は
$$
F_0 = S_0(1+r)^T
$$
です。現物を今買って満期まで資金コストを乗せる「キャリー」の考え方に対応します。

**補足:**
連続複利なら $F_0 = S_0 e^{rT}$ になります。ここでは選択肢の形に合わせて年複利を採用します。

```python
# 設問3: 理論先渡価格 F0 = S0 * (1 + r)^T の計算例
S0 = 100.0   # 現在のスポット価格
r = 0.02     # 固定金利（年複利）
T = 3        # 満期（年）

F0 = S0 * (1 + r) ** T

print(f"S0={S0}, r={r}, T={T}")
print(f"理論先渡価格 F0 = S0*(1+r)^T = {F0:.6f}")

# 検算: 割引現在価値は S0 に一致する（無裁定条件）
pv_forward_price = F0 / (1 + r) ** T
print(f"F0 の現在価値 = {pv_forward_price:.6f} (S0 と一致)")
```

### 設問４

$ t = 0 $先渡契約を締結した場合、$ t = 0 $時点での契約価格はいくらになりますか？つまり$ f_{0} $はいくらになりますか

1. 判断できません
2. 0
3. $ frac{S_{0}}{d(0, t)} $
4. $ S_{0} $



### 回答４

**正解:** 2. 0

**説明:**
先渡契約を $t=0$ で新規に締結する場合、受渡価格 $K$ は無裁定条件を満たすように設定されるため、契約価値は
$$
f_0 = 0
$$
になります。

**補足:**
このときの理論受渡価格は（配当等がない単純形では）
$$
K = \frac{S_0}{d(0,T)}
$$
です。つまり「価格 $K$」はゼロではなく、**契約価値 $f_0$ がゼロ**です。

```python
# 設問4: 契約締結時価値 f0 が 0 になることを数値確認
S0 = 100.0
r = 0.03
T = 2

# 割引因子 d(0,T) = 1/(1+r)^T（年複利）
d_0T = 1 / (1 + r) ** T

# 無裁定な受渡価格 K = S0 / d(0,T)
K = S0 / d_0T

# t=0 における先渡契約価値（ロング）
# f0 = S0 - K * d(0,T)
f0 = S0 - K * d_0T

print(f"S0={S0}, r={r}, T={T}")
print(f"d(0,T)={d_0T:.6f}")
print(f"理論受渡価格 K={K:.6f}")
print(f"契約価値 f0 = S0 - K*d(0,T) = {f0:.12f}")
```

# 金利の期間構造


### 設問１

ある宝くじで当選者に1000万ドルを支払うと宣言しています。
この賞金が500000ドルづつ（最初の支払いは即時）、合計20回支払われます。
この賞金を年利10%で複利計算した場合の現在価値はいくらになりますか？

回答は100万ドル単位で、小数点第二位を四捨五入してください。
(例)5.7124万ドルと計算した場合答えは5,71となります





### 回答１

**解答:** 現在価値は **約4.68百万ドル** です。

**説明:**
支払いは「即時の1回 + その後19回」の合計20回なので、年利10%複利での現在価値は
$$
PV = 500{,}000 \sum_{k=0}^{19}\frac{1}{(1+0.10)^k}
$$
で計算します。これは年金現価（初回即時払いなので年金前払）です。

**補足:**
問題文の指定に合わせて百万ドル単位で表すと、$4.6829$ 百万ドル $\Rightarrow 4.68$ 百万ドル（小数第3位を四捨五入）です。

```python
# 設問1 計算: 500,000ドルを即時1回 + 年末に19回、年利10%複利で割引
payment = 500_000
r = 0.10
n = 20

pv = sum(payment / (1 + r) ** k for k in range(n))  # k=0 が即時支払い
pv_million = pv / 1_000_000

print(f"現在価値 PV = ${pv:,.2f}")
print(f"百万ドル表示 = {pv_million:.4f} -> 四捨五入で {pv_million:.2f}")
```

### 設問２
ある夫婦が、6か月アパートを借りるために最初の1か月分の家賃1000ドルを手付金として支払いました。
この手付金は6か月後に返金されます。

翌日、彼らは同じように気に入ったアパートを見つけますが、そのアパートの月々の家賃は900ドルしかありません。
そしてまた、6か月後に返金される900ドルの補償金を入れなければなりません。

彼らはこのアパートに6か月しか住まない予定です。
新しいアパートに帰るべきでしょうか？金利を12%/月と仮定し、毎月複利で計算します。




### 回答２

**解答:** 前提を通常的に置くと、**新しいアパートへ移るべき**です。

**説明:**
現在時点からの比較で、毎月家賃は $1000 \to 900$ となるため、6か月で家賃差は有利です。さらに保証金（手付金）はどちらも6か月後返金なので、時間価値を考慮しても新アパート側の総現在価値コストが小さくなります。

**補足:**
手付金の返金条件（違約金有無・返金タイミング）が問題文で簡略化されているため、ここでは「両案とも6か月後に返金される」として現在価値比較を行っています。

```python
# 設問2 計算: 2案の現在価値コスト比較（仮定: 両案とも保証金は6か月後に返金）
i = 0.12   # 月利12%（月複利）
months = 6

rent_old = 1000
rent_new = 900
dep_old = 1000
dep_new = 900

# 月末払いの家賃現在価値
pv_rent_old = sum(rent_old / (1 + i) ** t for t in range(1, months + 1))
pv_rent_new = sum(rent_new / (1 + i) ** t for t in range(1, months + 1))

# 保証金のネット現在価値コスト = 今払う保証金 - 6か月後返金の現在価値
pv_dep_old = dep_old - dep_old / (1 + i) ** months
pv_dep_new = dep_new - dep_new / (1 + i) ** months

pv_total_old = pv_rent_old + pv_dep_old
pv_total_new = pv_rent_new + pv_dep_new

print(f"旧アパート総PVコスト: ${pv_total_old:,.2f}")
print(f"新アパート総PVコスト: ${pv_total_new:,.2f}")
print(f"差額(旧-新): ${pv_total_old - pv_total_new:,.2f}")
print("結論: 差額が正なら新アパートが有利")
```

### 設問３
1年物、2年物スポット金利が年複利で$ s_{1} = 6.3 \% $、$ s_{2} = 6.9 \%  $です。
※このコースでは特に指定がない場合、金利は常に年ベースで提示される頃を思い出して下さい

割引率$ d(0, 2) $はいくらですか？

回答は小数点第一位を四捨五入したパーセンテージで提出して下さい。



### 回答３

**解答:** 割引率 $d(0,2)$ は **約87.5%** です。

**説明:**
2年スポット金利 $s_2=6.9\%$（年複利）より
$$
d(0,2)=\frac{1}{(1+s_2)^2}=\frac{1}{(1.069)^2}\approx 0.8751
$$
したがってパーセンテージ表示では約 $87.5\%$ です。

**補足:**
1年スポット金利 $s_1$ はこの設問では不要で、2年の割引率には $s_2$ のみを使います。

```python
# 設問3 計算: d(0,2) = 1 / (1 + s2)^2
s2 = 0.069

d_02 = 1 / (1 + s2) ** 2

print(f"d(0,2) = {d_02:.6f}")
print(f"パーセンテージ表示 = {100*d_02:.1f}%")
```

### 設問４
1年物、2年物スポット金利が年複利で$ s_{1} = 6.3 \% $、$ s_{2} = 6.9 \%  $です。
※このコースでは特に指定がない場合、金利は常に年ベースで提示される頃を思い出して下さい

年複利と仮定した場合フォワードレート$ f_{1,2} $はいくらですか？

回答は小数点第一位を四捨五入したパーセンテージで提出して下さい。



### 回答４

**解答:** フォワードレート $f_{1,2}$ は **約7.5%** です。

**説明:**
年複利では
$$
(1+s_2)^2=(1+s_1)(1+f_{1,2})
$$
より
$$
f_{1,2}=\frac{(1+s_2)^2}{1+s_1}-1
=\frac{1.069^2}{1.063}-1\approx 0.0749
$$
なので約 $7.5\%$ です。

**補足:**
これは「1年後から2年後までの1年間」に対応するインプライド金利です。

```python
# 設問4 計算: f_{1,2} = (1+s2)^2/(1+s1) - 1
s1 = 0.063
s2 = 0.069

f_12 = (1 + s2) ** 2 / (1 + s1) - 1

print(f"f_{{1,2}} = {f_12:.6f}")
print(f"パーセンテージ表示 = {100*f_12:.1f}%")
```

### 設問５

現在の株価は1株当たり400ドルで配当はありません。
年率8％（四半期複利）で一定の金利が支払われると仮定した場合、9か月後の受け渡される株式の理論的な先渡し価格はいくらですか？

回答は小数点以下第2位を四捨五入して下さい。

### 回答５

**解答:** 理論先渡し価格は **424.48ドル** です。

**説明:**
配当なし、年率8%を四半期複利とすると、1期（3か月）金利は $0.08/4=0.02$。9か月は3期なので
$$
F_0=S_0\left(1+\frac{0.08}{4}\right)^3
=400\times 1.02^3
=424.4832
$$
より、四捨五入して $424.48$ ドルです。

**補足:**
連続複利でなく「四半期複利」を使う点が重要です。

```python
# 設問5 計算: F0 = S0 * (1 + r/m)^(mT)
S0 = 400
r = 0.08
m = 4        # 四半期複利
T = 0.75     # 9か月
n = int(m * T)

F0 = S0 * (1 + r / m) ** n

print(f"理論先渡し価格 F0 = ${F0:.4f}")
print(f"小数第3位を四捨五入: ${F0:.2f}")
```

### 設問６

借入金利が$ r_{B} = 10 \% $複利であるとします。

しかし、貸出金利（または預金金利）が年複利8%だけです。

年間$ A = 10,000 $ドルを支払う永久年金の価格の上限と下限を計算しなさい。

回答は1ドル未満を四捨五入して下さい。


### 回答６

**解答:** 永久年金価格の
- **下限:** 100,000ドル
- **上限:** 125,000ドル

**説明:**
年額 $A=10{,}000$ の永久年金は、割引率 $r$ に対して $P=\frac{A}{r}$。
借入金利 $r_B=10\%$、貸出金利 $r_L=8\%$ のとき、無裁定の価格帯は
$$
\frac{A}{r_B} \le P \le \frac{A}{r_L}
$$
なので
$$
\frac{10{,}000}{0.10}=100{,}000,\quad \frac{10{,}000}{0.08}=125{,}000
$$
です。

**補足:**
下限を下回ると借入して買う裁定、上限を上回ると売って貸し付ける裁定が可能になります。

```python
# 設問6 計算: 永久年金価格の下限・上限
A = 10_000
r_B = 0.10  # 借入金利
r_L = 0.08  # 貸出金利

P_lower = A / r_B
P_upper = A / r_L

print(f"下限価格 = ${P_lower:,.0f}")
print(f"上限価格 = ${P_upper:,.0f}")
```

### 設問７

今月から6か月後に満期を迎える株式の先渡し契約を保有しているとします。

この契約を締結したのは8か月前からなので、契約締結時の有効期限は$ T = 1 $年でした。

6か月前の株価は$ S_{0} = 100 $で現在の株価は125、現在の金利は$ r = 10 \% $（これは6か月前と同じ金利です）です。

為替予約の現在の価格はいくらですか？

ドル単位で1ドル未満を四捨五入して解答して下さい



### 回答７

**解答:** 先渡し契約の現在価値は **約20ドル（ロング）** です。

**説明:**
契約締結時（8か月前）に、満期1年の理論受渡価格は
$$
K=100(1+0.10)^1=110
$$
現在は満期まで6か月（$\tau=0.5$年）なので、ロング先渡しの価値は
$$
f_t=S_t-\frac{K}{(1+r)^{\tau}}
=125-\frac{110}{1.1^{0.5}}
\approx 20.12
$$
したがって四捨五入して約20ドルです。

**補足:**
契約締結時に固定された受渡価格 $K$ は変わらず、現在はその将来支払額を残存期間で割り引いて評価します。

```python
# 設問7 計算: 既存ロング先渡しの現在価値 f_t = S_t - K/(1+r)^tau
S_init = 100
S_t = 125
r = 0.10
T_original = 1.0
tau = 0.5   # 残存期間6か月

K = S_init * (1 + r) ** T_original
f_t = S_t - K / (1 + r) ** tau

print(f"締結時の受渡価格 K = {K:.4f}")
print(f"現在価値 f_t = {f_t:.4f}")
print(f"1ドル未満四捨五入 = {round(f_t):.0f} ドル")
```
