## 11.2 基本的な仮設検定
### 11.2.1 正規分布の母平均の検定(母分散既知)

<!-- auto-explain -->
### 正規分布にもとづく臨界値・p値計算

このセルでは標準化統計量と正規分布の分位点を使って判定境界またはp値を求めます。

モデル式:
$$Z=\frac{\bar{X}-\mu_0}{\sigma/\sqrt{n}}\sim\mathcal{N}(0,1)$$

```python
def pmean_test(sample, mean0, p_var, alpha=0.05):
    s_mean = np.mean(sample)
    n = len(sample)
    rv = stats.norm()
    interval = rv.interval(1 - alpha)

    z = (s_mean - mean0) / np.sqrt(p_var / n)
    if interval[0] <= z < interval[1]:
        print('帰無仮説を採択')
    else:
        print('帰無仮説を棄却')
    if z < 0:
        p = rv.cdf(z) * 2
    else:
        p = (1 - rv.cdf(z)) * 2
    print(f'p値は{p:.3f}')
```

<!-- auto-explain -->
### このセルの計算とモデル

このセルではサンプルから統計量を計算し、検定または推定に利用します。

モデル式:
$$T = \frac{\hat{\theta}-\theta_0}{SE(\hat{\theta})}$$

```python
pmean_test(sample, 130, 9)
```

### 11.2.2 正規分布の母分散の検定

<!-- auto-explain -->
### 正規分布にもとづく臨界値・p値計算

このセルでは標準化統計量と正規分布の分位点を使って判定境界またはp値を求めます。

モデル式:
$$Z=\frac{\bar{X}-\mu_0}{\sigma/\sqrt{n}}\sim\mathcal{N}(0,1)$$

```python
def pvar_test(sample, var0, alpha=0.05):
    u_var = np.var(sample, ddof=1)
    n = len(sample)
    rv = stats.chi2(n - 1)
    interval = rv.interval(1 - alpha)

    y = (n - 1) * u_var / var0
    if interval[0] <= y < interval[1]:
        print('帰無仮説を採択')
    else:
        print('帰無仮説を棄却')
    if y < rv.isf(0.5):
        p = rv.cdf(y) * 2
    else:
        p = (1 - rv.cdf(y)) * 2
    print(f'p値は{p:.3f}')
```

<!-- auto-explain -->
### 分散またはカイ二乗統計量の計算

このセルでは分散に関する統計量を計算し、母分散の検定や適合度評価に使います。

モデル式:
$$Y=\frac{(n-1)S^2}{\sigma_0^2} \sim \chi^2_{n-1}$$

```python
pvar_test(sample, 9)
```

### 11.2.3 正規分布の母平均検定(母分散未知)

<!-- auto-explain -->
### 正規分布にもとづく臨界値・p値計算

このセルでは標準化統計量と正規分布の分位点を使って判定境界またはp値を求めます。

モデル式:
$$Z=\frac{\bar{X}-\mu_0}{\sigma/\sqrt{n}}\sim\mathcal{N}(0,1)$$

```python
def pmean_test(sample, mean0, alpha=0.05):
    s_mean = np.mean(sample)
    u_var = np.var(sample, ddof=1)
    n = len(sample)
    rv = stats.t(n - 1)
    interval = rv.interval(1 - alpha)

    t = (s_mean - mean0) / np.sqrt(u_var / n)
    if interval[0] <= t < interval[1]:
        print('帰無仮説を採択')
    else:
        print('帰無仮説を棄却')
    if t < 0:
        p = rv.cdf(t) * 2
    else:
        p = (1 - rv.cdf(t)) * 2
    print(f'p値は{p:.3f}')
```

<!-- auto-explain -->
### このセルの計算とモデル

このセルではサンプルから統計量を計算し、検定または推定に利用します。

モデル式:
$$T = \frac{\hat{\theta}-\theta_0}{SE(\hat{\theta})}$$

```python
pmean_test(sample, 130)
```
